# Iris Flower Classification

```python
#Import Libraries
import numpy as np  #for arrays and photos
import pandas as pd #data manipulation
import matplotlib.pyplot as plt  #data visualization
import seaborn as sns  #statistical plotting 


%matplotlib inline
```

```python
#Import Dataset
cols = ['Sepal_Length_in_cm','Sepal_Width_in_cm', 'Petal_Length_in_cm','Petal_Width_in_cm','Species']

df = pd.read_csv('Iris.csv',names=cols,skiprows=[0])
```

```python
# Draw Lineplot of Sepal Length V/S Sepal Width
sns.lineplot(x="Sepal_Length_in_cm", y="Sepal_Width_in_cm", data=df)
  
# Giving title to the plot using Matplotlib
plt.title('Sepal Length v/s Sepal Width')
```

```python
df.describe()
```

```python
#Visualize the whole dataset
x_var = ["Sepal_Length_in_cm", "Sepal_Width_in_cm"]
y_var = ["Petal_Length_in_cm", "Petal_Width_in_cm"]
sns.set(style="ticks")
sns.pairplot(df, hue='Species', x_vars=x_var, y_vars=y_var, height=3.5, aspect=1.25)
plt.show()
```

```python
# #Drop the first row as it contains non-numeric labels
# df = df.iloc[1:, :]

# Convert to NumPy arrays
data = df.values
X = data[:, 0:4].astype(float)  # Features (sepal length, sepal width, petal length, petal width)
Y = data[:, 4]    # Target variable (species labels)
```

```python
print(X)
```

```python
print(Y)
```

```python
# Calculate average of each features
Y_Data = np.array([np.average(X[:, i][Y==j].astype('float32')) 
for i in range (X.shape[1])
    for j in (np.unique(Y))])
Y_Data_reshaped = Y_Data.reshape(4, 3)
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0, 1)
X_axis = np.arange(len(cols)-1)
width = 0.25
```

```python
# Plot the average
plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Virginica')
plt.xticks(X_axis, cols[:4])
plt.xlabel("Features of Flower")
plt.ylabel("Value in cm.")
plt.legend(bbox_to_anchor=(1.75,1))
plt.show()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X,Y, test_size=0.2, random_state=108)
```

```python
from sklearn.svm import SVC 
model_svc = SVC()
model_svc.fit(x_train,y_train) #model fitting
```

```python
predictionSVC = model_svc.predict(x_test)
#To calculate accuracy use accuracy_score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictionSVC)*100)
accuracySVC = accuracy_score(y_test,predictionSVC)
```

```python
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
#Prediction of the species from the input vector
prediction = model_svc.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

model_LR=LogisticRegression(solver='lbfgs',max_iter=1000)
model_LR.fit(x_train,y_train)
```

```python
predictionLR=model_LR.predict(x_test)

#To calculate accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictionLR)*100)
accuracyLR = accuracy_score(y_test,predictionLR)
```

```python
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
#Prediction of the species from the input vector
prediction2 = model_LR.predict(X_new)
print("Prediction of Species: {}".format(prediction2))
```

```python
#Decision Classifiers
from sklearn.tree import DecisionTreeClassifier

model_DTC=DecisionTreeClassifier()
model_DTC.fit(x_train,y_train)
```

```python
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
#Prediction of the species from the input vector
predictionDTC = model_DTC.predict(X_new)
print("Prediction of Species: {}".format(predictionDTC))
```

```python
import matplotlib.pyplot as plt

# Model Names
models = ['Model_SVC', 'Model_LR', 'Model_DTC']

# Accuracy
accuracies = [accuracySVC,accuracyLR ,accuracyDTC ]

# Bar chart
plt.bar(models, accuracies, color=['Red', 'Blue', 'Green'])

# Add labels and title
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison of Three Models')

# Display chart
plt.ylim(0, 1)  
plt.show()
```
